```python
import boto3
from load_app import lambda_handler_load
from like_app import lambda_handler_like

s3 = boto3.client("s3")
BUCKET_NAME = "preprod-cymonix-internal"
BASE_FOLDER = "import/multifile_schema/"

def get_event(job_id, org_id):
    event = {
        "Records": [
            {
                "eventVersion": "2.1",
                "eventSource": "aws:s3",
                "awsRegion": "us-west-2",
                "eventTime": "2025-06-26T11:50:41.814Z",
                "eventName": "ObjectCreated:Put",
                "userIdentity": {
                    "principalId": "AWS:AROA23TBNJUNRHJHNT2WN:cymonix-preprod"
                },
                "requestParameters": {
                    "sourceIPAddress": "10.192.20.150"
                },
                "responseElements": {
                    "x-amz-request-id": "E2RB54PVS6ZB8CJD",
                    "x-amz-id-2": "/uHCSk9s71QlroyOa4g+qpad7WPAQOftYw9hKOWFZRFGEkzC5DLoG9vQlRMJ7emSL9JIywwN6jEcSlQ2ozjKEc0+J35pblz+"
                },
                "s3": {
                    "s3SchemaVersion": "1.0",
                    "configurationId": "YjNjZDEzNDctOTczOS00Yzc4LWE0ZjQtZWQ3ODkzZTk4YzQ4",
                    "bucket": {
                        "name": "preprod-cymonix-internal",
                        "ownerIdentity": {
                            "principalId": "A3LDDQRJ5N5RCF"
                        },
                        "arn": "arn:aws:s3:::preprod-cymonix-internal"
                    },
                    "object": {
                        "key": f"import/multifile_schema/{org_id}/{job_id}/schema.json",
                        "size": 15742,
                        "eTag": "ebbc2094e4f1d97f1728f9151f7dde51",
                        "versionId": "yegtxkxdI6Zxo8u2BfpE6oDteZcM74Ec",
                        "sequencer": "00685D3411B2603C60"
                    }
                }
            }
        ]
    }
    return event

def get_message(job_id, org_id):
    message = f"import/izzy_liked/{job_id}/{org_id}/2332.json"
    return message

def list_folders(prefix):
    """List folders inside a given prefix (only immediate subdirectories)."""
    result = s3.list_objects_v2(Bucket=BUCKET_NAME, Prefix=prefix, Delimiter='/')
    return [content["Prefix"] for content in result.get("CommonPrefixes", [])]

def process_batch(batch):
    """
    Process a batch of (org_id, job_id) pairs.
    You can choose to run these in parallel if needed.
    """
    for org_id, job_id in batch:
        try:
            print(f"Processing org_id: {org_id}, job_id: {job_id}")
            event = get_event(job_id, org_id)
            res = lambda_handler_load(event, " ")
            print(f"res from load {res}")
            message = get_message(job_id, org_id)
            res1 = lambda_handler_like(message, " ")
            print(f"res from like {res1}")
        except Exception as e:
            print(f"Error processing org_id: {org_id}, job_id: {job_id}: {e}")

def fetch_all_json_files(org_ids=None, batch_size=100):
    """
    Fetch and process JSON schema files for specified orgs and jobs automatically.
    
    Args:
        org_ids (list): List of organization IDs to process. If None, processes all orgs.
        batch_size (int): Number of jobs to process in each batch.
    """
    all_json_data = {}
    batch = []
    
    # Get all organization folders
    org_folders = list_folders(BASE_FOLDER)
    
    for org_folder in org_folders:
        org_id = org_folder.rstrip('/').split('/')[-1]
        
        # Skip if specific org_ids are provided and this org_id is not in the list
        if org_ids is not None and org_id not in org_ids:
            print(f"Skipping org_id: {org_id} (not in specified list)")
            continue
            
        print(f"Processing organization: {org_id}")
        all_json_data[org_id] = {}
        
        job_folders = list_folders(org_folder)
        for job_folder in job_folders:
            job_id = job_folder.rstrip('/').split('/')[-1]
            
            # Optional: skip specific job_id
            if job_id == "00428df0-46a0-4f1b-a360-2b0ed0284891":
                print(f"Skipping org_id: {org_id} and job_id: {job_id}")
                continue
                
            print(f"Queueing org_id: {org_id}, job_id: {job_id}")
            batch.append((org_id, job_id))
            
            if len(batch) >= batch_size:
                print(f"Processing batch of {len(batch)} models")
                process_batch(batch)
                batch = []
        
        # Process any remaining jobs for this org_id
        if batch:
            print(f"Processing final batch of {len(batch)} models for org_id {org_id}")
            process_batch(batch)
            batch = []
    
    return "Data loaded successfully"

if __name__ == "__main__":
    # Example usage with specific org_ids
    org_ids_to_process = ['9','94','95','96','97','98','99']  # Add your desired org_ids here
    
    # To process only specific org_ids:
    result = fetch_all_json_files(org_ids=org_ids_to_process, batch_size=100)
    
    # To process all org_ids (original behavior):
    # result = fetch_all_json_files(batch_size=100)
    
    print(result)
```
